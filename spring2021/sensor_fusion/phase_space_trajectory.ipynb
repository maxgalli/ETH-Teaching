```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#plt.style.use(['seaborn-whitegrid'])
import math
from scipy.optimize import curve_fit
```

```python
# Remember to delete first row from the file
input_file = "data/1_surface.csv"
```

```python
def shift(arr, shift):
    """ Function to constrain the measurements in degrees inside a 0-360 range
    (it basically avoid the visual effect we get due to the arbitrary assignment 
    of 0 and 360 in the app)
    """
    res = []
    for v in arr:
        v += shift
        if v > 360:
            v -= 360
        if v < 0:
            v = 360 - abs(v)
        res.append(v)
    return np.array(res)
```

```python
df = pd.read_csv(input_file, sep=";")
```

```python
# Plot theta vs theta_dot

x = shift(df['ORIENTATION Z (azimuth °)'], -100)
y = df['GYROSCOPE Z (rad/s)']

fig, ax = plt.subplots(constrained_layout=True)
ax.set_title('Phase Space Trajectories')
ax.set_xlabel('ORIENTATION Z (azimuth °)')
ax.set_ylabel('GYROSCOPE Z (rad/s)')
plt.plot(x, y, marker='.', linestyle='', color='b')
plt.show();
```
