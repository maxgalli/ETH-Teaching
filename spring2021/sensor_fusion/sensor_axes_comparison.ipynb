```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# File names

surface = 'data/1_surface.csv'
long = 'data/2_longside.csv'
short = 'data/3_shortside.csv'
```

```python
df_su = pd.read_csv(surface, sep=';')
df_l = pd.read_csv(long, sep=';')
df_sh = pd.read_csv(short, sep=';')
```

```python
df_su.columns
```

### Angle

```python
fig, ax = plt.subplots(constrained_layout=True)
ax.set_title('Angle - Surface')
ax.set_xlabel('Time since start in ms ')
ax.set_ylabel('B projections')
plt.plot(df_su['Time since start in ms '], df_su['MAGNETIC FIELD X (μT)'], linestyle='solid', color='b')
plt.plot(df_su['Time since start in ms '], df_su['MAGNETIC FIELD Y (μT)'], linestyle='dashed', color='r')
plt.plot(df_su['Time since start in ms '], df_su['MAGNETIC FIELD Z (μT)'], linestyle='dotted', color='g')
plt.show();
```

```python
fig, ax = plt.subplots(constrained_layout=True)
ax.set_title('Angle - Long Side')
ax.set_xlabel('Time since start in ms ')
ax.set_ylabel('B projections')
plt.plot(
    df_l['Time since start in ms '][df_l['Time since start in ms '] < 10000], 
    df_l['MAGNETIC FIELD X (μT)'][df_l['Time since start in ms '] < 10000], 
    linestyle='solid', color='b'
)
plt.plot(
    df_l['Time since start in ms '][df_l['Time since start in ms '] < 10000], 
    df_l['MAGNETIC FIELD Y (μT)'][df_l['Time since start in ms '] < 10000], 
    linestyle='dashed', color='r'
)
plt.plot(
    df_l['Time since start in ms '][df_l['Time since start in ms '] < 10000], 
    df_l['MAGNETIC FIELD Z (μT)'][df_l['Time since start in ms '] < 10000], 
    linestyle='dotted', color='g'
)
plt.show();
```

```python
fig, ax = plt.subplots(constrained_layout=True)
ax.set_title('Angle - Short Side')
ax.set_xlabel('Time since start in ms ')
ax.set_ylabel('B projections')
plt.plot(df_sh['Time since start in ms '], df_sh['MAGNETIC FIELD X (μT)'], linestyle='solid', color='b')
plt.plot(df_sh['Time since start in ms '], df_sh['MAGNETIC FIELD Y (μT)'], linestyle='dashed', color='r')
plt.plot(df_sh['Time since start in ms '], df_sh['MAGNETIC FIELD Z (μT)'], linestyle='dotted', color='g')
plt.show();
```

### Rotation Speed

```python
fig, ax = plt.subplots(constrained_layout=True)
ax.set_title('Rotational Speed - Surface')
ax.set_xlabel('Time since start in ms ')
ax.set_ylabel('Rotational speed')
plt.plot(df_su['Time since start in ms '], df_su['GYROSCOPE X (rad/s)'], linestyle='solid', color='b')
plt.plot(df_su['Time since start in ms '], df_su['GYROSCOPE Y (rad/s)'], linestyle='dashed', color='r')
plt.plot(df_su['Time since start in ms '], df_su['GYROSCOPE Z (rad/s)'], linestyle='dotted', color='g')
plt.show();
```

```python
fig, ax = plt.subplots(constrained_layout=True)
ax.set_title('Rotational Speed - Long Side')
ax.set_xlabel('Time since start in ms ')
ax.set_ylabel('Rotational speed')
plt.plot(
    df_l['Time since start in ms '][df_l['Time since start in ms '] < 10000], 
    df_l['GYROSCOPE X (rad/s)'][df_l['Time since start in ms '] < 10000], 
    linestyle='solid', color='b'
)
plt.plot(
    df_l['Time since start in ms '][df_l['Time since start in ms '] < 10000], 
    df_l['GYROSCOPE Y (rad/s)'][df_l['Time since start in ms '] < 10000], 
    linestyle='dashed', color='r'
)
plt.plot(
    df_l['Time since start in ms '][df_l['Time since start in ms '] < 10000], 
    df_l['GYROSCOPE Z (rad/s)'][df_l['Time since start in ms '] < 10000], 
    linestyle='dotted', color='g'
)
plt.show();
```

```python
fig, ax = plt.subplots(constrained_layout=True)
ax.set_title('Rotational Speed - Short Side')
ax.set_xlabel('Time since start in ms ')
ax.set_ylabel('Rotational speed')
plt.plot(df_sh['Time since start in ms '], df_sh['GYROSCOPE X (rad/s)'], linestyle='solid', color='b')
plt.plot(df_sh['Time since start in ms '], df_sh['GYROSCOPE Y (rad/s)'], linestyle='dashed', color='r')
plt.plot(df_sh['Time since start in ms '], df_sh['GYROSCOPE Z (rad/s)'], linestyle='dotted', color='g')
plt.show();
```
